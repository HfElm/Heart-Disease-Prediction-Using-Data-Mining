## Partie 3 : DATA PREPARATION

Nous allons maintenant prÃ©parer les donnÃ©es pour lâ€™analyse :  
- Supprimer les colonnes inutiles  
- GÃ©rer les valeurs manquantes et aberrantes  
- Encoder les variables catÃ©gorielles  
- CrÃ©er la variable cible binaire

```python
# Afficher le nombre de valeurs manquantes
df.isnull().sum()
```

```python
# Supprimer les colonnes avec trop de valeurs manquantes (ca, thal, slope)
df = df.drop(columns=['ca', 'thal', 'slope'])
```

```python
# Supprimer les lignes contenant des valeurs manquantes restantes
df = df.dropna()
```

```python
# Remplacer les zÃ©ros incohÃ©rents par NaN dans 'trestbps' et 'chol'
df['trestbps'] = df['trestbps'].replace(0, np.nan)
df['chol'] = df['chol'].replace(0, np.nan)
```

```python
# Supprimer les lignes avec ces valeurs aberrantes
df = df.dropna(subset=['trestbps', 'chol'])
```

```python
# Supprimer les valeurs nÃ©gatives de oldpeak
df = df[df['oldpeak'] >= 0]
```

```python
#### Encodage des variables catÃ©gorielles

from sklearn.preprocessing import LabelEncoder

# Colonnes catÃ©gorielles Ã  encoder
cat_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'dataset']

le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

```

```python
# CrÃ©er une variable cible binaire (maladie = 1 si num > 0, sinon 0)
df['target'] = df['num'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Supprimer les colonnes inutiles
df = df.drop(columns=['num'])  # 'num' est remplacÃ©e par 'target'
```

# Partie 6

## Construction des ModÃ¨les

Nous allons maintenant entraÃ®ner plusieurs algorithmes sur nos donnÃ©es normalisÃ©es. Les modÃ¨les choisis sont :

- Random Forest
- RÃ©gression Logistique
- Support Vector Machine (SVM)

Chaque modÃ¨le est justifiÃ© par sa capacitÃ© Ã  traiter les donnÃ©es tabulaires, classer efficacement et gÃ©rer les corrÃ©lations entre variables.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# PrÃ©paration des donnÃ©es
X = df.drop('target', axis=1)
y = df['target']

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Division train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ModÃ¨les
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(random_state=42)
}

# EntraÃ®nement
for name, model in models.items():
    model.fit(X_train, y_train)
    print(f"{name} entraÃ®nÃ©.")
```

# Partie 7

## Ã‰valuation des ModÃ¨les

Chaque modÃ¨le est Ã©valuÃ© sur les mÃ©triques suivantes :

- Accuracy (Exactitude)
- Precision (PrÃ©cision)
- Recall (Rappel)
- F1 Score

Ces mesures permettent de comparer objectivement les performances des algorithmes.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Ã‰valuation
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"\n--> RÃ©sultats pour : {name}")
    print("Accuracy :", accuracy_score(y_test, y_pred))
    print("Precision :", precision_score(y_test, y_pred))
    print("Recall :", recall_score(y_test, y_pred))
    print("F1 Score :", f1_score(y_test, y_pred))
```

# Partie 8

## Discussion des RÃ©sultats et AmÃ©liorations

- **Random Forest** a gÃ©nÃ©ralement offert les meilleures performances globales.
- **SVM** peut Ãªtre sensible Ã  la normalisation, mais reste robuste.
- **RÃ©gression Logistique** est simple et interprÃ©table mais peut sous-performer sur des relations complexes.

###  Suggestions d'amÃ©lioration :

- Optimisation des hyperparamÃ¨tres (grid search)
- RÃ©duction de dimension ou sÃ©lection de variables
- Validation croisÃ©e

Nous allons illustrer cela avec un tuning de Random Forest.


```python
from sklearn.model_selection import GridSearchCV

# Tuning de Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
}

grid = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Meilleurs paramÃ¨tres :", grid.best_params_)

# RÃ©Ã©valuation
best_rf = grid.best_estimator_
y_pred_rf = best_rf.predict(X_test)
print("\n--> RÃ©sultats du meilleur Random Forest")
print("Accuracy :", accuracy_score(y_test, y_pred_rf))
print("Precision :", precision_score(y_test, y_pred_rf))
print("Recall :", recall_score(y_test, y_pred_rf))
print("F1 Score :", f1_score(y_test, y_pred_rf))
```
