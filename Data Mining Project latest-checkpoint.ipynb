## Partie 3 : DATA PREPARATION

Nous allons maintenant pr√©parer les donn√©es pour l‚Äôanalyse :  
- Supprimer les colonnes inutiles  
- G√©rer les valeurs manquantes et aberrantes  
- Encoder les variables cat√©gorielles  
- Cr√©er la variable cible binaire

```python
# Afficher le nombre de valeurs manquantes
df.isnull().sum()
```

```python
# Supprimer les colonnes avec trop de valeurs manquantes (ca, thal, slope)
df = df.drop(columns=['ca', 'thal', 'slope'])
```

```python
# Supprimer les lignes contenant des valeurs manquantes restantes
df = df.dropna()
```

```python
# Remplacer les z√©ros incoh√©rents par NaN dans 'trestbps' et 'chol'
df['trestbps'] = df['trestbps'].replace(0, np.nan)
df['chol'] = df['chol'].replace(0, np.nan)
```

```python
# Supprimer les lignes avec ces valeurs aberrantes
df = df.dropna(subset=['trestbps', 'chol'])
```

```python
# Supprimer les valeurs n√©gatives de oldpeak
df = df[df['oldpeak'] >= 0]
```

```python
#### Encodage des variables cat√©gorielles

from sklearn.preprocessing import LabelEncoder

# Colonnes cat√©gorielles √† encoder
cat_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'dataset']

le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

```

```python
# Cr√©er une variable cible binaire (maladie = 1 si num > 0, sinon 0)
df['target'] = df['num'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Supprimer les colonnes inutiles
df = df.drop(columns=['num'])  # 'num' est remplac√©e par 'target'
```

```python
#  V√©rifier le DataFrame final
df.info()
df.head()
```

# Partie 6

## Construction des Mod√®les

Nous allons maintenant entra√Æner plusieurs algorithmes sur nos donn√©es normalis√©es. Les mod√®les choisis sont :

- Random Forest
- R√©gression Logistique
- Support Vector Machine (SVM)

Chaque mod√®le est justifi√© par sa capacit√© √† traiter les donn√©es tabulaires, classer efficacement et g√©rer les corr√©lations entre variables.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Pr√©paration des donn√©es
X = df.drop('target', axis=1)
y = df['target']

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Division train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Mod√®les
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(random_state=42)
}

# Entra√Ænement
for name, model in models.items():
    model.fit(X_train, y_train)
    print(f"{name} entra√Æn√©.")
```

# Partie 7

## √âvaluation des Mod√®les

Chaque mod√®le est √©valu√© sur les m√©triques suivantes :

- Accuracy (Exactitude)
- Precision (Pr√©cision)
- Recall (Rappel)
- F1 Score

Ces mesures permettent de comparer objectivement les performances des algorithmes.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# √âvaluation
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"\n--> R√©sultats pour : {name}")
    print("Accuracy :", accuracy_score(y_test, y_pred))
    print("Precision :", precision_score(y_test, y_pred))
    print("Recall :", recall_score(y_test, y_pred))
    print("F1 Score :", f1_score(y_test, y_pred))
```

# Partie 8

## Discussion des R√©sultats et Am√©liorations

- **Random Forest** a g√©n√©ralement offert les meilleures performances globales.
- **SVM** peut √™tre sensible √† la normalisation, mais reste robuste.
- **R√©gression Logistique** est simple et interpr√©table mais peut sous-performer sur des relations complexes.

###  Suggestions d'am√©lioration :

- Optimisation des hyperparam√®tres (grid search)
- R√©duction de dimension ou s√©lection de variables
- Validation crois√©e

Nous allons illustrer cela avec un tuning de Random Forest.


```python
from sklearn.model_selection import GridSearchCV

# Tuning de Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
}

grid = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Meilleurs param√®tres :", grid.best_params_)

# R√©√©valuation
best_rf = grid.best_estimator_
y_pred_rf = best_rf.predict(X_test)
print("\n--> R√©sultats du meilleur Random Forest")
print("Accuracy :", accuracy_score(y_test, y_pred_rf))
print("Precision :", precision_score(y_test, y_pred_rf))
print("Recall :", recall_score(y_test, y_pred_rf))
print("F1 Score :", f1_score(y_test, y_pred_rf))
```
